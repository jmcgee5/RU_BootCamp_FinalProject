```python
import pandas as pd 

data = pd.read_csv('datasets/movies_final.csv')

urls = []

dList = data['imdb_id'].tolist()

def build_url(y):
    return 'https://www.imdb.com/title/' + str(y) + '/plotsummary?ref_=tt_stry_pl#synopsis'

for y in dList:
    url = build_url(y)
    urls.append(url)
```

```python
urls[7705] = 'https://www.imdb.com/title/tt1325753/plotsummary?ref_=tt_stry_pl#synopsis'
urls[10259] = 'https://www.imdb.com/title/tt0491603/plotsummary?ref_=tt_stry_pl#synopsis'

synopses = []
```

```python
import requests
from bs4 import BeautifulSoup
from html.parser import HTMLParser

for url in urls: 
    page_link = url
    page_response = requests.get(page_link, timeout=5)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    synopsis = page_content.find(id='plot-synopsis-content').text
    if len(synopsis) == 198:
        synopses.append('N/A')
    else: 
        synopses.append(synopsis)
        
print(synopses[0])
```

```python
data['synopsis'] = synopses
```

```python
data.to_csv('datasets/movies_final.csv', index=False)
```
